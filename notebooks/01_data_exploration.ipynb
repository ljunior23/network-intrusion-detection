
# Data Exploration



```python
# ============================================================================
# PART 1: IMPORTS AND SETUP
# ============================================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from datasets import load_dataset
from sklearn.preprocessing import LabelEncoder
from collections import Counter

warnings.filterwarnings('ignore')
sns.set_style('darkgrid')
plt.rcParams['figure.figsize'] = (12, 6)

print("Libraries loaded successfully!")
print("Using Hugging Face datasets for professional ML workflow")

# ============================================================================
# PART 2: LOAD DATASET FROM HUGGING FACE
# ============================================================================

print("\n" + "="*80)
print("LOADING NSL-KDD DATASET FROM HUGGING FACE")
print("="*80)

print("\nDownloading NSL-KDD from Hugging Face...")
print("This may take a minute on first run (cached afterwards)...\n")

try:
    dataset = load_dataset("Mireu-Lab/NSL-KDD")
    df_train = pd.DataFrame(dataset['train'])
    df_test = pd.DataFrame(dataset['test'])
    df = df_train.copy()
    
    print(f"âœ“ Dataset loaded successfully!")
    print(f"  Training set: {len(df_train):,} samples")
    print(f"  Test set: {len(df_test):,} samples")
    print(f"  Features: {df.shape[1]} columns")
    print(f"\nâœ“ Data cached locally for future use")
    
except Exception as e:
    print(f"âš  Error: {e}")
    raise

print(f"\nDataset Structure:")
print(f"Columns: {list(df.columns)[:10]}... (showing first 10)")

# ============================================================================
# PART 3: INITIAL DATA INSPECTION
# ============================================================================

print("\n" + "="*80)
print("DATASET OVERVIEW")
print("="*80)

display(df.head())
print("\nDataset Info:")
print(df.info())
print("\nBasic Statistics:")
display(df.describe())

print("\nMissing Values:")
missing = df.isnull().sum()
if missing.sum() > 0:
    display(missing[missing > 0])
else:
    print("âœ“ No missing values found")

# ============================================================================
# PART 4: TARGET VARIABLE ANALYSIS
# ============================================================================

print("\n" + "="*80)
print("ATTACK TYPE DISTRIBUTION")
print("="*80)

# Find target column
target_cols = ['label', 'attack_type', 'class', 'Label']
target_col = None
for col in target_cols:
    if col in df.columns:
        target_col = col
        break

print(f"Target column: {target_col}")
print("\nUnique attack types:")
display(df[target_col].value_counts())

# Create binary classification
df['is_attack'] = df[target_col].apply(
    lambda x: 0 if 'normal' in str(x).lower() else 1
)

print(f"\nClass Distribution:")
print(f"Normal Traffic: {(df['is_attack']==0).sum():,} ({(df['is_attack']==0).sum()/len(df)*100:.1f}%)")
print(f"Attack Traffic: {(df['is_attack']==1).sum():,} ({(df['is_attack']==1).sum()/len(df)*100:.1f}%)")

# Map attacks to categories
attack_map = {}
for val in df[target_col].unique():
    val_lower = str(val).lower()
    if 'normal' in val_lower:
        attack_map[val] = 'Normal'
    elif any(x in val_lower for x in ['dos', 'neptune', 'smurf', 'pod', 'teardrop', 'land', 'back']):
        attack_map[val] = 'DoS'
    elif any(x in val_lower for x in ['probe', 'portsweep', 'ipsweep', 'nmap', 'satan']):
        attack_map[val] = 'Probe'
    elif any(x in val_lower for x in ['r2l', 'guess_passwd', 'ftp_write', 'imap', 'phf', 'multihop', 'warezmaster', 'warezclient', 'spy']):
        attack_map[val] = 'R2L'
    elif any(x in val_lower for x in ['u2r', 'buffer_overflow', 'loadmodule', 'perl', 'rootkit']):
        attack_map[val] = 'U2R'
    else:
        attack_map[val] = 'Other'

df['attack_category'] = df[target_col].map(attack_map)
print("\nAttack Categories:")
display(df['attack_category'].value_counts())

# Visualize
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

df['is_attack'].value_counts().plot(kind='bar', ax=axes[0], color=['green', 'red'])
axes[0].set_title('Normal vs Attack Traffic', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Traffic Type')
axes[0].set_ylabel('Count')
axes[0].set_xticklabels(['Normal', 'Attack'], rotation=0)

df['attack_category'].value_counts().plot(kind='bar', ax=axes[1], color='coral')
axes[1].set_title('Attack Categories Distribution', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Attack Category')
axes[1].set_ylabel('Count')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# ============================================================================
# PART 5: FEATURE ANALYSIS
# ============================================================================

print("\n" + "="*80)
print("FEATURE ANALYSIS")
print("="*80)

numeric_features = df.select_dtypes(include=[np.number]).columns.tolist()
numeric_features = [f for f in numeric_features if f not in ['is_attack']]
categorical_features = df.select_dtypes(include=['object']).columns.tolist()
categorical_features = [f for f in categorical_features if f not in [target_col, 'attack_category']]

print(f"\nNumeric Features: {len(numeric_features)}")
print(f"Categorical Features: {len(categorical_features)}")
print(f"\nCategorical: {categorical_features}")

for feature in categorical_features:
    print(f"\n{feature}:")
    display(df[feature].value_counts().head())

# ============================================================================
# PART 6: CORRELATION ANALYSIS
# ============================================================================

print("\n" + "="*80)
print("CORRELATION ANALYSIS")
print("="*80)

# Select subset for correlation
selected_features = numeric_features[:15] + ['is_attack']
corr_matrix = df[selected_features].corr()

plt.figure(figsize=(14, 10))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, mask=mask, annot=True, fmt='.2f', cmap='coolwarm', 
            center=0, square=True, linewidths=1)
plt.title('Feature Correlation Heatmap', fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()

target_corr = corr_matrix['is_attack'].sort_values(ascending=False)
print("\nTop 10 features correlated with attacks:")
display(target_corr.head(10))

# ============================================================================
# PART 7: FEATURE DISTRIBUTIONS
# ============================================================================

print("\n" + "="*80)
print("FEATURE DISTRIBUTIONS")
print("="*80)

key_features = numeric_features[:8]

fig, axes = plt.subplots(2, 4, figsize=(18, 8))
axes = axes.ravel()

for idx, feature in enumerate(key_features):
    data_normal = df[df['is_attack']==0][feature]
    data_attack = df[df['is_attack']==1][feature]
    
    axes[idx].hist(data_normal, bins=50, alpha=0.6, label='Normal', color='green', density=True)
    axes[idx].hist(data_attack, bins=50, alpha=0.6, label='Attack', color='red', density=True)
    axes[idx].set_title(feature, fontweight='bold')
    axes[idx].legend()
    axes[idx].set_yscale('log')

plt.suptitle('Feature Distributions: Normal vs Attack Traffic', 
             fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

# ============================================================================
# PART 8: SAVE PROCESSED DATA (For Next Notebook)
# ============================================================================

print("\n" + "="*80)
print("SAVING DATA FOR FEATURE ENGINEERING")
print("="*80)

# Create output directory
import os
os.makedirs('../data/processed', exist_ok=True)

# Save training data with labels
df_train_labeled = df.copy()  # This is our training data with labels
df_train_labeled.to_csv('../data/processed/train_with_labels.csv', index=False)
print("âœ“ Saved: ../data/processed/train_with_labels.csv")

# Save test data with labels
df_test['is_attack'] = df_test[target_col].apply(
    lambda x: 0 if 'normal' in str(x).lower() else 1
)
df_test['attack_category'] = df_test[target_col].map(attack_map)
df_test.to_csv('../data/processed/test_with_labels.csv', index=False)
print("âœ“ Saved: ../data/processed/test_with_labels.csv")

# Save feature metadata
feature_info = {
    'numeric_features': numeric_features,
    'categorical_features': categorical_features,
    'target_column': target_col,
    'attack_categories': list(attack_map.values())
}

import json
with open('../data/processed/feature_info.json', 'w') as f:
    json.dump(feature_info, f, indent=4)
print("âœ“ Saved: ../data/processed/feature_info.json")

# ============================================================================
# SUMMARY
# ============================================================================

print("\n" + "="*80)
print("EDA SUMMARY REPORT")
print("="*80)

print(f"""
Dataset: NSL-KDD Network Intrusion Detection
Total Samples: {len(df):,}
Normal Traffic: {(df['is_attack']==0).sum():,} ({(df['is_attack']==0).sum()/len(df)*100:.1f}%)
Attack Traffic: {(df['is_attack']==1).sum():,} ({(df['is_attack']==1).sum()/len(df)*100:.1f}%)

Attack Categories:
{df['attack_category'].value_counts().to_string()}

Features:
- Numeric: {len(numeric_features)}
- Categorical: {len(categorical_features)}

Key Insights:
1. Dataset is imbalanced - need to address in modeling
2. Strong correlations found between certain features and attacks
3. Clear separation in feature distributions between normal and attack traffic
4. Ready for feature engineering and model training

Next Steps:
â†’ Feature engineering and selection
â†’ Handle class imbalance with SMOTE
â†’ Train multiple ML models
â†’ Build neural network with GPU acceleration
""")

print("\n" + "="*80)
print("EDA COMPLETE! Ready for Feature Engineering")
print("="*80)
```
